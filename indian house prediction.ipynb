```python
def tier(city):
    if city in tier1:
        return 0
    elif city in tier2:
        return 1
    else:
        return 2;
    
    
    
Data["CITY_TIER"]=Data["ADDRESS"].apply(tier)
# Data.drop("city_tier",axis=1,inplace=True)
```

```python
Data["POSTED_BY"]=Data["POSTED_BY"].astype("category")
Data["POSTED_BY"]=Data["POSTED_BY"].cat.codes
Data["BHK_OR_RK"]=Data["BHK_OR_RK"].astype("category")
Data["BHK_OR_RK"]=Data["BHK_OR_RK"].cat.codes
```
